# Polynomial Regression
 - Special case of linear regression model useful for curvilinear relationships
 curvilinear relationships
 - By squaring or setting higher order forms for pedictor variables
 Before polynomial regression we have to preprocess data .
We have to change degree of data and then build the model.

- if we changed degree more, it will overfit


```python
x = df['highway-mpg']
y = df['price']
```

```python
import numpy as np
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
PlotPolly(p,x,y,'highway-mpg')
```

```python
def PlotPolly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(15,55,100)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title("Poliynomial Fit")
    ax = plt.gca()
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.show()
    plt.close()
```

```python
def PlotPolly1(model, Name):
    x_new = np.linspace(15,55,100)
    y_new = model(x_new)
    
    plt.plot(x, y, '.')
    plt.title("Polynomial Fit")
    ax = plt.gca()
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.show()
    plt.close()
```

```python
import numpy as np
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
PlotPolly1(p,'highway-mpg')
```

```python
import numpy as np
f = np.polyfit(x, y, 6)
p = np.poly1d(f)
print(p)
PlotPolly(p,x,y,'highway-mpg')
```
